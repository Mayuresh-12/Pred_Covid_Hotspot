```python
def get_final_counts_from_df(df, key, typeval, sortkey):
    fips = df[key].astype(typeval).unique()
    l = []
    for each in fips:
        d = df[df[key] == each].sort_values(by=sortkey)
        if len(d) > 0:
            l.append(d.iloc[-1])
        else:
            print("Error ", d)
    print(len(l))
    s = pd.DataFrame(l)
    return s
d = get_final_counts_from_df(dataframe, "FIPS", "int", "Date")
```
